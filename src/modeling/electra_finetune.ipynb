```python
from tqdm.notebook import tqdm
test_data = [test.iloc[idx, 0] for idx in range(len(test))]
test_pred = []
# predict
max_len = 128
for idx, review in tqdm(enumerate(test_data)):
  pred = sentiment_classifier(review[:200])
  test_pred.append(pred[0]['label'])
  print(f'{review}\n>> {pred[0]}')
```

```python
label_dict = {'LABEL_0' : '하락', 'LABEL_1' : '중립', 'LABEL_2' : '상승', 'LABEL_3' : '매우하락', 'LABEL_4' : '매우상승'}
test['pred'] = test_pred
test['pred'] = test['pred'].apply(lambda x : label_dict[x])
test.head()
```

```python
eval_list = []
for idx in test.index:
  if test.loc[idx, 'target'] == test.loc[idx, 'pred']:
    eval_list.append(True)
  else :
    eval_list.append(False)
test['eval'] = eval_list
test.head()
```
