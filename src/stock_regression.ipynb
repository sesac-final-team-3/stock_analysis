```python
import pandas as pd
import os
```

```python
df1 = pd.read_csv('./../Data/stock_data/final_kospi_OHLCV.csv')
df2 = pd.read_csv('./../Data/stock_data/final_kospi_foreign_investment.csv')
df3 = pd.read_csv('./../Data/stock_data/final_kospi_fundmental.csv')
df4 = pd.read_csv('./../Data/stock_data/final_kospi_shorting_balance.csv')
df5 = pd.read_csv('./../Data/stock_data/final_kospi_trading_value.csv')
df6 = pd.read_csv('./../Data/stock_data/final_kospi_trading_volume.csv')
```

```python
df1['code'] = df1['code'].apply(lambda x : str(x).zfill(6))
df2['code'] = df2['code'].apply(lambda x : str(x).zfill(6))
df3['code'] = df3['code'].apply(lambda x : str(x).zfill(6))
df4['code'] = df4['code'].apply(lambda x : str(x).zfill(6))
df5['code'] = df5['code'].apply(lambda x : str(x).zfill(6))
df6['code'] = df6['code'].apply(lambda x : str(x).zfill(6))
```

```python
df1 = df1.loc[df1['code']=='000270']
df2 = df2.loc[df2['code']=='000270']
df3 = df3.loc[df3['code']=='000270']
df4 = df4.loc[df4['code']=='000270']
df5 = df5.loc[df5['code']=='000270']
df6 = df6.loc[df6['code']=='000270']
```

```python
new_df = pd.merge(df1[['날짜','거래대금','등락률']],df4[['날짜','공매도금액']],how='inner',on='날짜')
```

```python
new_df = pd.merge(new_df,df5[['날짜', '기관합계', '기타법인', '개인', '외국인합계']],how='inner',on='날짜')
```

```python
new_df = pd.merge(new_df,df6[['날짜', '기관합계', '기타법인', '개인', '외국인합계']],how='inner',on='날짜')
```

```python
new_df = pd.merge(new_df,df2[['날짜','지분율']],how='inner',on='날짜')
```

```python
new_df
```

```python
x = new_df.drop(['날짜','등락률'],axis=1)
y = new_df['등락률']
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x, y)
```

```python
for i in range(len(x.columns)):
    if i != (len(x.columns)-1):
        print(model.coef_[i],'*',x.columns[i],'+')
    else:
        print(model.coef_[i],'*',x.columns[i])
```

```python
import statsmodels.api as sm

model1 = sm.OLS(y, sm.add_constant(x)).fit()
model1.summary()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scale = scaler.fit_transform(x)
x_scaled = pd.DataFrame(data=x_scale, columns=x.columns)
```

```python
import statsmodels.api as sm

model2 = sm.OLS(y, sm.add_constant(x_scaled)).fit()
model2.summary()
```
