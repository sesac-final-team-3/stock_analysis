# KOSPI

```python
from pykrx import stock
# list
kospi_tickers = stock.get_market_ticker_list(market="KOSPI")
```

```python
import time
import pandas as pd
import os
temp=[]
for ticker in kospi_tickers:
    kospi_ticker_df = stock.get_exhaustion_rates_of_foreign_investment("20120101", "20221226", ticker)
    kospi_ticker_df.reset_index(inplace=True)
    kospi_ticker_df['code'] = str(ticker)
    temp.append(kospi_ticker_df)
    print(ticker,'의 외국인 투자정보를 입력완료 ')
    print('쌓인 종목 수',len(temp))
    
    time.sleep(1)
```

```python
final_kospi=pd.concat(temp,axis=0)
final_kospi.reset_index(inplace=True)
final_kospi.to_csv('../../../../workspace/data/krx/final_kospi_foreign_investment.csv',index=None,encoding="utf-8-sig")
```

### Load

```python
kospi_f_df=pd.read_csv('../../../../workspace/data/krx/final_kospi_foreign_investment.csv',index_col=None)
```

```python
kospi_f_df=kospi_f_df.drop(kospi_f_df.columns[0],axis=1)
kospi_f_df['code']=kospi_f_df['code'].apply(lambda x : str(x).zfill(6))
```

# Kosdaq

```python
from pykrx import stock
# list
kosdaq_tickers = stock.get_market_ticker_list(market="KOSDAQ")
```

```python
import time
import pandas as pd
import os
temp=[]
empty_ticker_list=[]
for ticker in kosdaq_tickers:
    try:
        kosdaq_ticker_df = stock.get_exhaustion_rates_of_foreign_investment("20120101", "20221226", ticker)
        kosdaq_ticker_df.reset_index(inplace=True)
        kosdaq_ticker_df['code'] = str(ticker)
        temp.append(kosdaq_ticker_df)
        print(ticker,'의 외국인 투자정보를 입력완료 ')
        print('쌓인 종목 수',len(temp))
        
        time.sleep(1)
    except:
        empty_ticker_list.append(ticker)
        print(ticker)
```

```python
final_kosdaq=pd.concat(temp,axis=0)
final_kosdaq.reset_index(inplace=True)
final_kosdaq.to_csv('../../../../workspace/data/krx/final_kosdaq_foreign_investment.csv',index=None,encoding="utf-8-sig")
```
