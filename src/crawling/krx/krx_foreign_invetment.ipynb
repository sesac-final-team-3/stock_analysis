```python
from pykrx import stock
# list
kospi_tickers = stock.get_market_ticker_list(market="KOSPI")
kosdaq_tickers = stock.get_market_ticker_list(market="KOSDAQ")
# naming
kospi_name=[stock.get_market_ticker_name(ticker) for ticker in kospi_tickers]
kosdaq_name=[stock.get_market_ticker_name(ticker) for ticker in kosdaq_tickers]
```

```python
# 예시
kospi_ticker_df = stock.get_exhaustion_rates_of_foreign_investment("20221220", "20221226", '005930')
kospi_ticker_df.reset_index(inplace=True)
kospi_ticker_df['code'] = '005930'
kospi_ticker_df
```

```python
import time
import pandas as pd
import os
temp=[]
for ticker in kospi_tickers:
    kospi_ticker_df = stock.get_exhaustion_rates_of_foreign_investment("20120101", "20221226", ticker)
    kospi_ticker_df.reset_index(inplace=True)
    kospi_ticker_df['code'] = ticker
    temp.append(kospi_ticker_df)
    print(ticker,'의 외국인 투자정보를 입력완료 ')
    print('쌓인 종목 수',len(temp))
    
    time.sleep(1)
```

```python
final_kospi=pd.concat(temp,axis=0)
final_kospi.reset_index(inplace=True)
final_kospi.to_csv('./final_kospi_foreign_investment.csv',index=False,encoding="utf-8-sig")
```
