```python
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36'}

for i in range(1,11):
    url = f'https://m.stock.naver.com/api/index/KPI200/enrollStocks?pageSize=20&page={i}&type=object'
    response = requests.get(url,headers=header)
    for i in range(len(response.json()['stocks'])):
        stock_name = response.json()['stocks'][i]['stockName']
        stock_code = response.json()['stocks'][i]['itemCode']
        df = df.append({'company':stock_name, 'code':stock_code},ignore_index=True)
```
