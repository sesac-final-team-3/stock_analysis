```python
import pandas as pd
import requests

test = pd.read_xml('CORPCODE.xml')
test
```

```python
# 005380 현대차 / 035420 NAVER / 068270 셀트리온 / 005930 삼성전자 / 000270 기아 / 
# 373220 LG에너지솔루션 / 207940 삼성바이오로직스 / 000660 SK하이닉스 / 006400 삼성SDI / 051910 LG화학
kospi_list = ["005380","035420","068270","005930","000270","373220","207940","000660","006400","051910"]
```

```python
name_list = []

for i in kospi_list:
    temp = test.loc[test['stock_code']==int(i)]
    code = temp['corp_code'].values[0]
    company = temp['corp_name'].values[0]
    print(company,f'{code:08d}')
    crtfc_key = crtfc_key # dart api key
    corp_code = f'{code:08d}'
    url = 'https://opendart.fss.or.kr/api/company.json?crtfc_key='+crtfc_key+'&corp_code=' + corp_code
    result=requests.get(url)
    print(result.json()['jurir_no'],'\n')

    name = {'stock_code':str(f'{int(i):06d}'), 'name':company, 'crno':result.json()['jurir_no']} # 'code':code, 

    name_list.append(name)

name_df = pd.DataFrame(name_list)
name_df
```

```python
name_df.to_csv('stock_code.csv', index=False)
```

```python
corporate_code_list=[]
company_list=[]

for i in kospi_list:
    temp = test.loc[test['stock_code']==int(i)]
    code = temp['corp_code'].values[0]
    company = temp['corp_name'].values[0]
    # print(company,f'{code:08d}')
    crtfc_key = crtfc_key # dart api key
    corp_code = f'{code:08d}'
    url = 'https://opendart.fss.or.kr/api/company.json?crtfc_key='+crtfc_key+'&corp_code=' + corp_code
    result=requests.get(url)
    company_list.append(company)
    corporate_code = result.json()['jurir_no']
    corporate_code_list.append(corporate_code)
```

```python
# 원래대로 검색하면 안떠서 따로 추가
company_list.append('에스케이하이닉스')
company_list.append('네이버')
company_list.append('엘지에너지솔루션')
```

```python
import requests
import json

serviceKey = serviceKey
# 20120101 / 20221226

# 기업 정보
# DB : code / established_date / listed_date / CEO / settlement_date / tel
# DB : 코드 / 설립일 / 상장일 / CEO / 결산월 / 전화번호
# API : (stock_code) / enpEstbDt / enpXchgLstgDt / enpRprFnm / enpStacMm / enpTlno

for crno in corporate_code_list:
    for company in company_list:
        try: 
            url = 'http://apis.data.go.kr/1160100/service/GetCorpBasicInfoService/getCorpOutline'
            params ={'serviceKey' : serviceKey, 'pageNo' : '1', 'numOfRows' : '5', 'resultType' : 'json', 'basDt' : '20221226', 'crno' : crno, 'corpNm' : company}

            response = requests.get(url, params=params)
            contents = response.text
            # print(contents)

            json_ob = json.loads(contents)
            # print(json_ob)
            # print(type(json_ob)) #json타입 확인

            body = json_ob['response']['body']['items']['item']
            # print(body)

            df = pd.json_normalize(body)
            df = df[['enpEstbDt', 'enpXchgLstgDt', 'enpRprFnm', 'enpStacMm', 'enpTlno', 'crno']]
            # print(df)

            df.to_csv('df_test_info.csv', mode='a') # mode='a' : 누적 

        except:
            pass
```

```python
df1 = pd.read_csv('stock_code.csv', dtype=str) # 종목코드때문에 dtype=str 로 가져옴
df2 = pd.read_csv('api_info.csv', dtype=str)
```

```python
df2
```

```python
df1['crno']=pd.to_numeric(df1['crno']) # Python int too large to convert to C long 오류 해결

df2['crno']=pd.to_numeric(df2['crno'])
```

```python
# crno(동일 컬럼) 기준으로 병합
df_merge = pd.merge(df1, df2, on = 'crno', how = 'outer')
df_merge
```

```python
# 컬럼 순서 변경
df_merge.columns.tolist()
df_merge = df_merge[['stock_code', 'name', 'enpEstbDt', 'enpXchgLstgDt', 'enpRprFnm', 'enpStacMm', 'enpTlno']] # crno 삭제
```

```python
# DB에 맞게 컬럼 이름 수정
# DB : code / established_date / listed_date / CEO / settlement_date / tel
# DB : 코드 / 설립일 / 상장일 / CEO / 결산월 / 전화번호
# API : / enpEstbDt / enpXchgLstgDt / enpRprFnm / enpStacMm / enpTlno

df_merge = df_merge.rename(columns={'stock_code':'code', 'enpEstbDt':'established_date', 
                        'enpXchgLstgDt':'listed_date', 'enpRprFnm':'CEO', 'enpStacMm':'settlement_date', 'enpTlno':'tel'})
df_merge
```

```python
# listed_date 형식 변경
df_merge['listed_date'] = df_merge['listed_date'].apply(lambda x: '-'.join(x.split('/')[::-1]))
df_merge['listed_date']
```

```python
# 주식 이름 drop
df_merge = df_merge.drop(['name'], axis=1, inplace=False)
df_merge
```

```python
df_merge['established_date'] = pd.to_datetime(df_merge['established_date'])
df_merge['listed_date'] = pd.to_datetime(df_merge['listed_date'])

df_merge
```

```python
df_merge.to_csv('info_final.csv', index=False)
```
