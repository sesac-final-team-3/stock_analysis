```python
import pandas as pd
```

```python
df1 = pd.read_csv('./../Data/kospi_report.csv')
df2 = pd.read_csv('./../Data/kosdaq_report.csv')
```

```python
new_df = pd.concat([df1,df2])
```

```python
new_df.reset_index(drop=True,inplace=True)
```

```python
new_df['coment'].value_counts()
```

```python
new_df['coment1'] = new_df['coment'].apply(lambda x: "Buy" if x =="매수" else 
                                         '??' if x =="MarketPerform" else
                                         '??' if x =="OutPerform" else
                                         '??' if x =="투자의견없음" else
                                         'Hold' if x =="중립" else
                                         'StrongBuy' if x =="강력매수" else
                                         '??' if x =="없음" else
                                         '??' if x =="시장수익률상회" else
                                         x)
```

```python
df_sort_group_top1 = new_df.sort_values(by="date", ascending=False).groupby("code").head(1)
```

```python
df_sort_group_top1.sort_values('code',ascending=False)['coment']
```

```python
df_sort_group_top1.sort_values('code',ascending=False)['coment'].value_counts()
```
